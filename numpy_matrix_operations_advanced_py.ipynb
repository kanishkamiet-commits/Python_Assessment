```python
# Program: NumPy Matrix Operations
# This program creates two matrices and performs:
# addition, subtraction, multiplication, and inverse operations

import numpy as np

# Create first matrix
matrix1 = np.array([
    [1, 2],
    [3, 4]
])

# Create second matrix
matrix2 = np.array([
    [5, 6],
    [7, 8]
])

# Display matrices
print("Matrix 1:\n", matrix1)
print("\nMatrix 2:\n", matrix2)

# Matrix Addition
addition = matrix1 + matrix2

# Matrix Subtraction
subtraction = matrix1 - matrix2

# Matrix Multiplication
multiplication = np.dot(matrix1, matrix2)

# Inverse of matrices
inverse_matrix1 = np.linalg.inv(matrix1)
inverse_matrix2 = np.linalg.inv(matrix2)

# Display results
print("\n--- Matrix Addition ---")
print(addition)

print("\n--- Matrix Subtraction ---")
print(subtraction)

print("\n--- Matrix Multiplication ---")
print(multiplication)

print("\n--- Inverse of Matrix 1 ---")
print(inverse_matrix1)

print("\n--- Inverse of Matrix 2 ---")
print(inverse_matrix2)
```
