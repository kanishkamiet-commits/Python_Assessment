```python
# Program: Pie Chart for Smartphone Market Share
# This program plots a pie chart showing market share
# of smartphone brands and highlights the highest market share

import matplotlib.pyplot as plt

# Smartphone brands
brands = ["Samsung", "Apple", "Xiaomi", "OnePlus", "Realme"]

# Market share percentages
market_share = [35, 30, 15, 10, 10]

# Explode the brand with maximum market share
explode_values = [0.1, 0, 0, 0, 0]

# Plot pie chart
plt.pie(
    market_share,
    labels=brands,
    explode=explode_values,
    autopct='%1.1f%%',
    shadow=True
)

# Add title
plt.title("Smartphone Market Share")

# Display chart
plt.show()
```
