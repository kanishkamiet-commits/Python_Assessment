```python
# Program: Read sales data from CSV and plot bar chart using Matplotlib
# This program uses Pandas for data handling and Matplotlib for visualization

import pandas as pd
import matplotlib.pyplot as plt

# Create sample CSV file (run once)
data = {
    "Product": ["Pen", "Book", "Pencil", "Marker", "Eraser"],
    "Sales": [120, 200, 150, 80, 60]
}

df_sample = pd.DataFrame(data)
df_sample.to_csv("sales.csv", index=False)

# Read CSV file
df = pd.read_csv("sales.csv")

# Display data
print("\nSales Data:\n")
print(df)

# Plot bar chart
plt.bar(df["Product"], df["Sales"], label="Product Sales")

# Add title and labels
plt.title("Product-wise Sales Data")
plt.xlabel("Products")
plt.ylabel("Sales")

# Add legend and grid
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```
