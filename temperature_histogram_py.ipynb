```python
# Program: Temperature Distribution Histogram
# This program generates random temperature data using NumPy
# and plots a histogram using Matplotlib

import numpy as np
import matplotlib.pyplot as plt

# Generate random temperature data
# Random temperatures between 20°C and 40°C
temperature_data = np.random.randint(20, 41, 100)

# Display generated data
print("Temperature Data:\n")
print(temperature_data)

# Plot histogram
plt.hist(temperature_data, bins=10)

# Add title and labels
plt.title("Temperature Distribution Histogram")
plt.xlabel("Temperature (°C)")
plt.ylabel("Frequency")

# Add grid
plt.grid(True)

# Display histogram
plt.show()
```
