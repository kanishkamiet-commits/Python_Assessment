```python
# Program: Line Chart for Monthly Sales Data
# This program uses Matplotlib to display monthly sales of a company
# with title, labels, legend, and grid for better visualization

import matplotlib.pyplot as plt

# Months of the year
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Monthly sales data (example values)
sales = [12000, 15000, 18000, 14000, 20000, 22000,
         21000, 23000, 25000, 24000, 26000, 30000]

# Plot line chart
plt.plot(months, sales, marker='o', label="Monthly Sales")

# Add title and labels
plt.title("Company Monthly Sales Data")
plt.xlabel("Months")
plt.ylabel("Sales (in INR)")

# Add grid and legend
plt.grid(True)
plt.legend()

# Display chart
plt.show()
```
