```python
# Program: Student Attendance Analysis using Pandas
# This program reads attendance records from a CSV file
# and displays students with attendance below 75%

import pandas as pd

# Create sample CSV file (run once)
data = {
    "Student_Name": ["Aman", "Riya", "John", "Sara", "Neha"],
    "Attendance_Percentage": [82, 68, 74, 90, 65]
}

df_sample = pd.DataFrame(data)
df_sample.to_csv("attendance.csv", index=False)

# Read CSV file
df = pd.read_csv("attendance.csv")

# Display full attendance data
print("\nStudent Attendance Records:\n")
print(df)

# Filter students with attendance below 75%
low_attendance = df[df["Attendance_Percentage"] < 75]

# Display result
print("\nStudents with Attendance Below 75%:\n")
print(low_attendance)
```
