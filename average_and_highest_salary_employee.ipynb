```python
# Program: Employee Data Analysis using Pandas
# This program reads employee data from a CSV file and calculates:
# department-wise average salary and highest salary employee

import pandas as pd

# Create sample CSV file (run once)
data = {
    "EmpID": [101, 102, 103, 104, 105, 106],
    "Name": ["Aman", "Riya", "John", "Sara", "Vikram", "Neha"],
    "Department": ["HR", "IT", "IT", "Finance", "HR", "Finance"],
    "Salary": [40000, 60000, 55000, 70000, 45000, 80000]
}

df_sample = pd.DataFrame(data)
df_sample.to_csv("employees.csv", index=False)

# Read CSV file
df = pd.read_csv("employees.csv")

# Display dataset
print("\nEmployee Data:\n")
print(df)

# Department-wise average salary
print("\nDepartment-wise Average Salary:\n")
avg_salary = df.groupby("Department")["Salary"].mean()
print(avg_salary)

# Find highest salary employee
highest_salary_employee = df.loc[df["Salary"].idxmax()]

print("\nHighest Salary Employee:")
print("EmpID:", highest_salary_employee["EmpID"])
print("Name:", highest_salary_employee["Name"])
print("Department:", highest_salary_employee["Department"])
print("Salary:", highest_salary_employee["Salary"])
```
