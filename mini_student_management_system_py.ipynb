```python
# Program: Mini Student Management System
# This program uses:
# functions, dictionaries, file handling,
# exception handling, and Pandas for report generation

import pandas as pd

# Dictionary to store student records
students = {}

# Function to add student
def add_student():
    try:
        roll_no = int(input("Enter Roll Number: "))
        name = input("Enter Student Name: ")
        marks = float(input("Enter Marks: "))

        students[roll_no] = {
            "Name": name,
            "Marks": marks
        }

        print("Student added successfully!")

    except ValueError:
        print("Invalid input! Please enter correct data types.")


# Function to display all students
def display_students():

    if len(students) == 0:
        print("No student records found.")
        return

    print("\n--- Student Records ---")

    for roll_no, details in students.items():
        print("Roll No:", roll_no)
        print("Name:", details["Name"])
        print("Marks:", details["Marks"])
        print()


# Function to save records to file
def save_to_file():

    try:
        file = open("students.txt", "w")

        for roll_no, details in students.items():
            record = f"{roll_no},{details['Name']},{details['Marks']}\n"
            file.write(record)

        file.close()

        print("Records saved to file successfully!")

    except Exception as e:
        print("Error while saving file:", e)


# Function to generate report using Pandas
def generate_report():

    try:
        # Read file data
        df = pd.read_csv(
            "students.txt",
            names=["Roll_No", "Name", "Marks"]
        )

        print("\n--- Student Report ---\n")
        print(df)

        # Display statistics
        print("\nAverage Marks:", round(df["Marks"].mean(), 2))
        print("Highest Marks:", df["Marks"].max())
        print("Lowest Marks:", df["Marks"].min())

    except FileNotFoundError:
        print("Student file not found!")

    except Exception as e:
        print("Error while generating report:", e)


# Menu-driven program
while True:

    print("\n--- Mini Student Management System ---")
    print("1. Add Student")
    print("2. Display Students")
    print("3. Save Records to File")
    print("4. Generate Report")
    print("5. Exit")

    try:
        choice = int(input("Enter your choice (1-5): "))

        if choice == 1:
            add_student()

        elif choice == 2:
            display_students()

        elif choice == 3:
            save_to_file()

        elif choice == 4:
            generate_report()

        elif choice == 5:
            print("Exiting Student Management System.")
            break

        else:
            print("Invalid choice! Please enter between 1 and 5.")

    except ValueError:
        print("Please enter a valid numeric choice.")
```
