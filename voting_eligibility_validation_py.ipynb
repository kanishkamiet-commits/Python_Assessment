```python
# Program: Voting Eligibility Checker with Exception Handling
# This program validates user age and checks voting eligibility
# while handling invalid numeric and non-numeric inputs

try:
    # Take age input from user
    age = int(input("Enter your age: "))

    # Validate negative age
    if age < 0:
        print("Invalid age! Age cannot be negative.")

    # Check voting eligibility
    elif age >= 18:
        print("You are eligible to vote.")

    else:
        print("You are not eligible to vote.")

# Handle non-numeric input
except ValueError:
    print("Error: Please enter a valid numeric age.")

# Handle unexpected errors
except Exception as e:
    print("Unexpected error occurred:", e)
```
