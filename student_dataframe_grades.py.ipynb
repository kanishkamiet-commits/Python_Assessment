```python
# Program: Student Marks Analysis using Pandas DataFrame
# This program calculates total marks, percentage,
# and assigns grades using apply() function

import pandas as pd

# Create DataFrame with student marks
data = {
    "Name": ["Aman", "Riya", "John", "Sara", "Neha"],
    "Math": [85, 92, 70, 88, 95],
    "Science": [90, 85, 75, 91, 89],
    "English": [78, 88, 80, 84, 93]
}

df = pd.DataFrame(data)

# Calculate total marks
df["Total"] = df[["Math", "Science", "English"]].sum(axis=1)

# Calculate percentage
df["Percentage"] = (df["Total"] / 300) * 100

# Function to assign grades
def assign_grade(percentage):
    if percentage >= 90:
        return "A+"
    elif percentage >= 80:
        return "A"
    elif percentage >= 70:
        return "B"
    elif percentage >= 60:
        return "C"
    else:
        return "Fail"

# Apply function to assign grades
df["Grade"] = df["Percentage"].apply(assign_grade)

# Display final DataFrame
print("\nStudent Result DataFrame:\n")
print(df)
```
